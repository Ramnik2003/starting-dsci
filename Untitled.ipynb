```r
library(tidyverse)
library(repr)
library(tidymodels)
library(readr)
library(dplyr)
library(rsample)
```

```r
# Read the dataset from the web into R
tennis_players <- read_csv("data/tennis_players.csv")
tournament_players <- tennis_players |>
mutate(winner_name=as.factor(winner_name))|>
select(winner_name, tourney_id, match_num)
tournament_players
# Preprocess data
```

```r
# Split the dataset into training and testing sets
library(rsample)
set.seed(123) # for reproducibility
data_split <- initial_split(tournament_players, prop = 0.8)
data_train <- training(data_split)
data_test <- testing(data_split)
data_train
data_test
```

```r
tennis_recipe <- recipe(winner_name ~ tourney_id + match_num, data=tournament_players)|>
step_center(all_predictors())|>
step_scale(all_predictors())
tennis_recipe
```

```r
knn_spec <- nearest_neighbor(weight_func ="rectangular", neighbors=tune())|>
set_engine("kknn")|>
set_mode("classification")
knn_spec
```

```r
tennis_workflow <- workflow() |>
  add_recipe(tennis_recipe) |>
  add_model(knn_spec)
tennis_workflow
```

```r
model <- fit(tennis_workflow, data_train)
```
